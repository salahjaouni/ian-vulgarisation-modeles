```python
!pip install transformers gradio

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
import gradio as gr
from datetime import datetime

# 🔁 Chargement du modèle T5 (taille moyenne, rapide)
model_name = "t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# 🧠 Fonction de génération (adaptée à la tâche)
def vulgariser_t5(texte, niveau):
    # Création du prompt selon le niveau
    if niveau == "Lycéen":
        instruction = "simplify:"
    elif niveau == "Étudiant":
        instruction = "rephrase for better understanding:"
    else:
        instruction = "explain in simple terms:"

    prompt = f"{instruction} {texte}"

    # Encodage et génération
    input_ids = tokenizer.encode(prompt, return_tensors="pt", max_length=512, truncation=True)
    output_ids = model.generate(input_ids, max_length=200, num_beams=4, early_stopping=True)
    result = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Sauvegarde
    enregistrer_t5(texte, result, niveau)
    return result

# 💾 Sauvegarde
def enregistrer_t5(texte, resultat, niveau):
    timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    nom_fichier = f"t5_vulgarisation_{niveau}_{timestamp}.txt"
    with open(nom_fichier, "w") as f:
        f.write("Texte original :\n")
        f.write(texte + "\n\n")
        f.write("Niveau : " + niveau + "\n\n")
        f.write("Texte vulgarisé :\n")
        f.write(resultat)

# 🎛️ Interface
interface_t5 = gr.Interface(
    fn=vulgariser_t5,
    inputs=[
        gr.Textbox(label="Texte scientifique à vulgariser", lines=8),
        gr.Radio(choices=["Lycéen", "Étudiant", "Grand public"], label="Public cible", value="Lycéen")
    ],
    outputs=gr.Textbox(label="Texte vulgarisé", lines=10),
    title="🧠 Vulgarisateur IA - T5",
    description="Génération open-source basée sur le modèle T5 (Google)."
)

# 🚀 Lancement
interface_t5.launch(share=True)
```
